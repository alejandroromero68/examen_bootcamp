```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la ecuación
def funcion(x):
    return x**2 - 8*x + 16

# Generar valores de x
x = np.linspace(0, 10, 100)  # Rango de valores de x, ajusta si es necesario

# Calcular los valores de y
y = funcion(x)

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$y = x^2 - 8x + 16$", color="blue")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Línea horizontal en y=0
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Línea vertical en x=0
plt.title("Gráfica de la ecuación $y = x^2 - 8x + 16$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
